```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
# get list of cover images and stego images
cover_images = sorted(os.listdir('data_test/cover'))
stego_images = sorted(os.listdir('data_test/stego'))
```

```python
# display 5 cover images and stego images
figures = []

for i in range(5):
    figures.append(plt.figure(figsize=(10, 10)))
    figures[i].add_subplot(1, 2, 1)
    plt.imshow(cv2.imread('data_test/cover/' + cover_images[i]))
    figures[i].add_subplot(1, 2, 2)
    plt.imshow(cv2.imread('data_test/stego/' + stego_images[i]))
    plt.show()
```
